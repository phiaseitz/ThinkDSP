```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy
import pandas
```

```python
nrows = 1625
df = pandas.read_csv('coindesk-bpi-USD-close_data-2015-02-07_2015-02-08.csv', nrows=nrows, parse_dates=[0])
ys = df.Close.values
#ys = mat[:,1]
wave = thinkdsp.Wave(ys, framerate=1)
wave.plot()
```

```python
def corrcoef(xs, ys):
    return numpy.corrcoef(xs, ys, ddof=0)[0, 1]

def serial_corr(wave, lag=1):
    N = len(wave)
    y1 = wave.ys[lag:]
    y2 = wave.ys[:N-lag]
    corr = corrcoef(y1, y2)
    return corr

def autocorr(wave):
    """Computes and plots the autocorrelation function.

    wave: Wave
    """
    lags = range(len(wave.ys)//2)
    corrs = [serial_corr(wave, lag) for lag in lags]
    return lags, corrs

lags,corrs = autocorr(wave)
thinkplot.plot(lags, corrs, label='Test')
```
