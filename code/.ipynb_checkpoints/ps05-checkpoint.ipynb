```python
from __future__ import print_function, division

import thinkdsp
import thinkplot
import thinkstats2
import math
import numpy

%precision 3
%matplotlib inline

PI2 = 2 * math.pi
i = complex(0, 1)
```

1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.

```python
def idft(amps):
    N = len(amps)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = (M.conj().transpose().dot(amps))/N
    return ys
```

```python
def dft(ys):
    N = len(ys)
    ts = numpy.arange(N) / N
    freqs = numpy.arange(N)
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = M.conj().transpose().dot(ys)
    return amps
```

```python
def idft_allen(amps):
    ys = dft(amps) / N
    return ys
```

```python
framerate = 10000
signal = thinkdsp.SawtoothSignal(freq=500)
wave = signal.make_wave(duration=0.1, framerate=framerate)
seg = wave.segment(0,0.01)
seg.plot()

```

```python
print(wave.ys)

hs = dft(wave.ys)
len(wave.ys), len(hs)

amps = numpy.absolute(hs)

N = len(hs)
fs = numpy.arange(N) * framerate / N

thinkplot.plot(fs[:N//2+1], amps[:N//2+1])
```

```python
iys = idft(amps)
print(iys)
iys1 = idft_allen(amps)
thinkplot.plot(iys[:N//2+1])
```
