Doing DCT Compression

On PS03, it looks like some people did not get as far as a working implementation of DCT-based compression.  If you want to finish it for this homework, I think you will find it educational, and you can fill in some of the missing points from PS03.

Split into segments
do the DCT 
get the n lowest amplitudes, and get rid of them
return that.

```python
def dctCompression(wave,amp_cutoff,num_segments):
    duration_segments = wave.duration/num_segments
    last_segment = wave.duration - duration_segments
    framerate = wave.framerate

    compressed_segments = []
    
    for start_time in numpy.linspace(0,last_segment,num_segments):
        segment = wave.segment(start_time, duration_segments)
        segment_dct = segment.make_dct()
        amps = segment_dct.amps
        
        keepind = numpy.where(amps > amp_cutoff)
        
        keepamps = numpy.zeros(amps.shape)
        keepamps[keepind] = 1
        
        amps2 = numpy.multiply(amps,keepamps)
             
        new_dct = thinkdsp.Dct(amps2,framerate)
        compressed_segment = new_dct.make_wave()
        compressed_segments.append(compressed_segment)
    
    return concatSegs(compressed_segments)

def concatSegs(seg_list):
    wave = seg_list[0]
    for seg in seg_list[1:]:
        seg.apodize()
        wave = wave | seg
    wave.normalize()
    wave.apodize()
    return wave.segment(0.5)
```

```python
compressed_wave = dctCompression(seg,30,500)


compressed_wave.plot()
compressed_wave.make_audio()
```

Question 1 - Shifting in time!

```python
def synthesize1(amps, freqs, ts):
    components = [thinkdsp.ComplexSinusoid(freq, amp)
                  for amp, freq in zip(amps, freqs)]
    signal = thinkdsp.SumSignal(*components)
    ys = signal.evaluate(ts)
    return ys



def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys
```

```python
amps = numpy.array([0.6, 0.25, 0.1, 0.05])
freqs = [100, 200, 300, 400]
framerate = 11025

ts = numpy.linspace(0, 1, framerate)
ys = synthesize1(amps, freqs, ts)
```

```python
def shiftAmps (amps, freqs, to):
    return numpy.multiply(amps,numpy.exp(i*to*numpy.array(freqs)))
```

```python
amps2  = shiftAmps(amps, freqs, 1)

ys2 = synthesize2(amps2,freqs,ts)
```

```python
n=400
thinkplot.plot(ts[:n], ys.real[:n], label=r'$Originial$')
thinkplot.plot(ts[:n], ys2.real[:n], label=r'$Shifted$')
thinkplot.config(ylim=[-1.15, 1.05])
```
