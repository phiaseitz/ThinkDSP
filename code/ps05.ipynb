1) Following the example in chap07.ipynb, make a synthesis matrix (the IDFT matrix) using a different range of frequencies, and see what difference it makes.

```python
def approx_equal(a, b, tol=1e-10):
    return sum(abs(a-b)) < tol
```

```python
#IDFT MATRIX
N = 8
ts = numpy.arange(N) / N
freqs = numpy.arange(N)
args = numpy.outer(ts, freqs)
M1 = numpy.exp(i * PI2 * args)
```

```python
N = 8
ts = numpy.arange(N) / N
freqs = numpy.arange(N,2*N)
args = numpy.outer(ts, freqs)
M2 = numpy.exp(i * PI2 * args)
```

```python
print(approx_equal(M1,M2))
```

2) In thinkdsp.py, the Spectrum methods low_pass and high_pass are ideal filters, in some sense: in the passband they are perfectly flat, and in the stopband they attenuate to 0 (or whatever value you want).

Invoking the principle that for every filter there is a corresponding window, see if you can figure out what window corresponds to these ideal filters.


```python
fs = numpy.linspace(0,1,101)
mid = 0.5
hpamps = numpy.zeros(101)
lpamps = numpy.zeros(101)

hpamps[numpy.where(fs > mid)] = 1
lpamps[numpy.where(fs < mid)] = 1

```

```python
def synthesize2(amps, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    ys = numpy.dot(M, amps)
    return ys
```

```python
ts = numpy.linspace(0,10,100)
hpys = synthesize2(hpamps, fs, ts)
lpys = synthesize2(lpamps, fs, ts)
```

```python
thinkplot.plot(ts, hpys.real, label=r'$High Pass$')
thinkplot.plot(ts, lpys.real, label=r'$Low Pass$')
thinkplot.config()
```

So far we have focused attention on the effect filters have on the magnitude of each frequency component, but they also affect the phase offset.  For the boxcar and Gaussian windows, plot the phase shift as a function of frequency.

```python
N = 101
ts = numpy.linspace(0,1,N)

boxcar = numpy.ones(N)
boxcar /= sum(boxcar)

gaussian = scipy.signal.gaussian(M=N, std=2)
gaussian /= sum(gaussian)
```

```python
def analyze1(ys, freqs, ts):
    args = numpy.outer(ts, freqs)
    M = numpy.exp(i * PI2 * args)
    amps = numpy.linalg.solve(M, ys)
    return amps
```

```python
freqs = numpy.linspace(0,10,101)

ampsboxcar = analyze1(boxcar,freqs,ts)
ampsgaussian = analyze1(gaussian,freqs,ts)

phaseboxcar = numpy.angle(ampsboxcar)
print(ampsboxcar)
print(phaseboxcar)
phasegaussian = numpy.angle(ampsgaussian)
print(ampsgaussian)
print(phasegaussian)
```

```python
thinkplot.plot(freqs, phaseboxcar, label=r'$Boxcar$')
thinkplot.plot(freqs, phasegaussian, label=r'$Gaussian$')
thinkplot.config()
```
